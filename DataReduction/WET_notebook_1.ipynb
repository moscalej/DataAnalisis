```python

#Q1.1
# section1

u = np.random.uniform(0, 20, 3000)
v = np.random.uniform(0, 1, 3000)
x = u * np.cos(u)
y = u * np.sin(u)
z = v

color = u / u.max()

```

```python
# section2
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z, c=color)
fig, ax = plt.subplots(1, 3)
ax[0].scatter(x, y, c=color)
ax[1].scatter(x, z, c=color)
ax[2].scatter(z, y, c=color)
plt.show()
```
